# **Exercise 11: Simulate Crashing of Python program**
* if we didn't formulate the code properly

```python
# Print some message
print("Give me two numbers and i will divide for you.")
print("Enter 'q' to quit")

# While loop to handle division
# Program will crash if second_number = 0
while True:
    first_num = input('\nFirst number: ')
    if first_num == 'q':
        break
    second_num = input('Second number: ')
    if second_num == 'q':
        break
    ans = int(first_num)/ int(second_num)
    print(ans)
```

# **Exercise 12: The try-except-else block**

```python
# Print some message
print("Give me two numbers and i will divide for you.")
print("Enter 'q' to quit")
# Use the try-except-else block
while True:
    first_num = input('\nFirst number: ')
    if first_num == 'q':
        break
    second_num = input('Second number: ')
    try:
        ans = int(first_num)/ int(second_num)
    except ZeroDivisionError:
        print("You can't divide by 0")
    else:
        print(ans)
```

# **Exercise 15: Using a function**

```python
# Define a count_words function
def count_words(filename):
    """Count the appropriate number of words in a file"""
    try:
        with open(filename) as f_obj:
            contents = f_obj.read()
    except FileNotFoundError:
        msg = "Sorry, the file " + filename + " does not exist."
        print(msg)
    else:
        #count the appropritae number of words in the file
        words = contents.split()
        num_words = len(words)
        print("The file " + filename + " is about " + str(num_words) + " words")
```

# **Exercise 17: Failing Silently**

```python
# Define a count_words function
# make some changes -> fail silently
def count_words(filename):
    """Count the appropriate number of words in a file"""
    try:
        with open(filename) as f_obj:
            contents = f_obj.read()
    except FileNotFoundError:
        pass # make failure become silently here
    else:
        #count the appropritae number of words in the file
        words = contents.split()
        num_words = len(words)
        print("The file " + filename + " is about " + str(num_words) + " words")
```

```python
# Lists of multiple files
filenames = ['alice.txt', 'first american king.txt','Jefferson.txt','azlin.txt']


# Count for words
# The siddhartha.txt -> Failing Silently
for filename in filenames:
    count_words(filename)
```

# **Exercise 21: Refactoring: improve the code by breaking it up into a series of functions that have specific jobs**
* makes our code

```python
# Import the json library
import json

# Define a greet_user() function
def greet_user():
    """Greet the user by name"""
    filename = "usernamesssss.json"
    try:
        with open(filename) as f_obj:
            username = json.load(f_obj)
    except FileNotFoundError:
        username = input("What is your name?")
        with open(filename,'w') as f_obj:
            json.dump(username, f_obj)
            print("We'll remember you when you come back " + username + "!")
    else:
        print("Welcome back, " + username + "!")
```

#**Refactoring of the function**

```python
# Load the json library
import json

# Refactoring the functions
def get_stored_username():
    """Get stored username if available."""
    filename = 'username.json'
    try:
        with open(filename) as f_obj:
            username = json.load(f_obj)
    except FileNotFoundError:
        return None
    else:
        return username

def get_new_username():
    """Prompt for a new username."""
    username = input("What is your name? ")
    filename = 'username.json'
    with open(filename, 'w') as f_obj:
        json.dump(username, f_obj)
    return username

def greet_user():
    """Greet the user by name."""
    username = get_stored_username()
    if username:
        print("Welcome back, " + username + "!")
    else:
        username = get_new_username()
        print("We'll remember you when you come back, " + username + "!")
```
