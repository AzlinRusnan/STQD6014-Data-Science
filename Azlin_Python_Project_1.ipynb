```python
# Coding #

# Upload and read the file
print("Uploading Mother's Remedies.txt")

# Read the Mother's Remedies.txt file
filename = "/content/Mother's Remedies.txt"

# Function to clean and extract words from the text
def extract_and_clean_words(text):
    """Remove non-alphabetic characters and split the text into lower case"""
    clean_text = ''
    for char in text:
        if char.isalpha() or char.isspace():
            clean_text += char.lower()

    words = clean_text.split()
    return words

# Function to count a specific keyword
def count_specific_keyword(keyword, words):
    """Count the occurences of a specific keyword."""
    keyword = keyword.lower()
    count = 0
    for word in words:
        if word == keyword:
            count += 1
    return count


# Function to find the top ten keywords
def top_ten_keywords(words):
    """Find and return top ten keyword"""
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    sorted_words = sorted(word_count, key=word_count.get, reverse=True)
    return sorted_words[:10], word_count

# Main analysis
try:
    with open(filename) as f_obj:
        contents = f_obj.read()
except FileNotFoundError:
    msg = "Sorry, the file " + filename + " does not exist."
    print(msg)
else:
    print("\nAnalyzing the file...")

    # Extract and clean words from the file contents
    words = extract_and_clean_words(contents)

    # Excludes common words that gives no meaning to the text, so that we can understand what the text is all about. I'm excluding more than common words that Dr Bernard suggested.
    common_words = set(['a', 'the', 'and', 'to', 'from', 'where', 'what', 'was', 'when', 'i', 'you', 'we', 'us',
                        'me', 'for', 'of','in','is','or','it','are','as','this','be','should','may','by','if',
                        'on','at','an','not','that','will','can','very','there','by','with','have','but','they','then','which','all','her','has'])
    filtered_words = []
    for word in words:
        if word not in common_words:
            filtered_words.append(word)

    # Count the total number of words excluding common words
    num_words = len(filtered_words)
    print("\nTotal Words (Excluding Common Words):", num_words)

    # Calculate the top ten keywords
    top_ten, word_count = top_ten_keywords(filtered_words)
    print("\nTop 10 Keywords:")
    for word in top_ten:
        count = word_count[word]
        percentage = (count/num_words)*100
        print("'" + word + "': Count = " + str(count) + ", Percentage = " + str(round(percentage, 2)) + "%")


    # User requested keyword count
    user_keyword = input("\nEnter the keyword you want to count: ")
    keyword_count = count_specific_keyword(user_keyword, filtered_words)
    print("The count of '" + user_keyword + "' in the text is: " + str(keyword_count))
```

For example, 'water' and 'milk' probably indicating with ingredients in the recipes or home remedies, which fits with the Mother's remedy theme.
3.   'one', 'two', and 'three' are among the top keywords, indicating that numerical quantitis are mentioned frequently, possibly related to dosages or recipe quantities.
